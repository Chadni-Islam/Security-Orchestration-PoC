```python
for new_api in new_api_list:
        found = 0
        max_similarity_score = 0
        similarity = float()
        for api in API_list:
            print(api, new_api)
            synsets_api = wordnet.synsets(api)
            # print(synsets_api, '\n')
            synsets_new_api = wordnet.synsets(new_api)
            # print(synsets_new_api, '\n')
            for s1 in synsets_api:
                for s2 in synsets_new_api:
                    # print(s1.name(), s2.name())
                    similarity = s1.wup_similarity(s2)
                    if not similarity:
                        continue
                    s1_pos = s1.pos()
                    s2_pos = s2.pos()
                    print(similarity)
                    if similarity > 0.6:
                        found = 1
                        if s1_pos == s2_pos:
                            print(new_api, ' similar to API: ', api)
                            # print(similarity, s1.name(), s2.name(), '\n')
                            if similarity > max_similarity_score:
                                max_similarity_score = similarity
                                max_similar_api = api
                        
            if found == 1:
                print('final: ', new_api, 'similar: t0 ', api,' ', max_similarity_score, max_similar_api)
            else:
                print('None')
        
```
